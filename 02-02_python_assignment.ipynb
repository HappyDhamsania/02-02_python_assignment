For loop is useful when no of iterations are not known.
when condition for breaking the loop is not specified.

```python
#printing 0 to 10
for i in range(11):
    print(i)
```

While loop is useful when the no of iterations are known.
when loop has to be breaked if a condition is fulfilled.

```python
#printing 0 to 10
i = 0
while i < 11:
    print(i)
    i=i+1
```

```python
#first 10 natural numbers sum using for loop
sum=0
for i in range(1,11):
    sum=sum+i
print(sum)
```

```python
#first 10 natural numbers product using for loop
product=1
for i in range(1,11):
    product=product*i
print(product)
```

```python
#first 10 natural numbers sum using while loop
s=1
c=2
while c< 11:
    s=s+c
    c=c+1
print(s)
```

```python
#first 10 natural numbers product using while loop
s=1
c=2
while c< 11:
    s=s*c
    c=c+1
print(s)
```

```python
units=int(input('enter units'))
bill=0
if units<=100:
    bill=bill+4.5*units
    units=units-100
elif units>100 and units<=200:
    bill=bill+450
    units=units-100
    bill=bill+6*units
elif units>200 and units<=300:
    bill=bill+450+600
    units=units-200
    bill=bill+10*units
elif units>300:
    bill=bill+450+600+1000
    units=units-300
    bill=bill+20*units
print(bill)
```

```python
#using for loop
h=list(range(1,101))
s=[]
for i in h:
    r=i**3
    if r%4==0:
        s.append(i)
    elif r%5==0:
        s.append(i)
print(s)
```

```python
#using while loop
s=[]
i=1
while i <= 100:
    r=i**3
    if r%4==0:
        s.append(i)
    elif r%5==0:
        s.append(i)
print(s)
```
